```python
import pandas as pd
```

```python
columns = ["Sex", "Length", "Diameter", "Height", "Whole_weight",
               "Shucked_weight", "Viscera_weight", "Shell_weight", "Rings"]
```

```python
path = "data/trial_env/data"
path_prepared = f"{path}/prepared_data"
```

```python
# load main dataset
data = pd.read_csv(f"{path}/abalone.csv", names=columns)
data
```

```python
# set the last 600 rows to make the reserve dataset
reserved_data = data.iloc[-600:, :]
reserved_data
```

```python
# Take out the last 600 rows: This will be used to build the model.
df = data.drop(reserved_data.index)
df
```

```python
# Save the training reserve and dataset to csv files
df.to_csv(f"{path_prepared}/abalone_data.csv", index=False)  # , index=False

reserved_data.to_csv(f"{path_prepared}/reserved_data.csv", index=False)
```

```python
df_iter = pd.read_csv(f"{path_prepared}/reserved_data.csv", iterator=True, chunksize=50)

n = 1


while True:
    try:
        df = next(df_iter)
        df.to_csv(f"{path_prepared}/reserved_{n}.csv", index=False)
        print(f"length of reserved_{n}.csv: {len(df)}")
        n += 1
    except StopIteration:
        print("Finished saving data to csv files")
        break
```
